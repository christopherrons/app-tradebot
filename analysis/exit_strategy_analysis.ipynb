```python
import numpy as np
import math
import plotly.graph_objects as go
```

```python
fig = go.Figure()
for initial_sell_quantity in [20, 50, 100]:
    quantity = 6250
    investment_cost = quantity * 0.2
    values = []
    prices = np.arange(0.20, 20, 0.1)
    profit = 0

    for idx, price in enumerate(prices):
        value = price * quantity
        values.append(value) 
        if price % 0.20 < math.pow(10, -5):
            sell_quantity = initial_sell_quantity
            while len(values) > 1 and value > prices[idx+1] * (quantity-sell_quantity):
                sell_quantity -= 1
            quantity -= sell_quantity
            profit += sell_quantity * price

    print(f"Final Quantity {quantity}")
    print(f"Price {price}")
    print(f"Final Value {value}")
    print(f"Profit {profit}")
    print(f"Value + Profit: {profit + value}")
    print("")
    fig.add_trace(go.Scatter(x=prices, y=values,
                    mode='lines',
                    name=f'Initial Sell Quantity: {initial_sell_quantity}'))
```

```python
fig.update_layout(title=f'Profit When selling every 0.20 cent gain',
                   xaxis_title='Price',
                   yaxis_title='Profit [$]')
fig.show()
```
